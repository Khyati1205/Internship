Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# set up Chrome driver
driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/")

# close login popup if it appears
try:
    driver.find_element(By.XPATH, '//button[text()="✕"]').click()
except:
    pass

# find search bar and search for smartphone
search_box = driver.find_element(By.NAME, "q")
search_box.send_keys("Oneplus Nord")
search_box.submit()

# wait for search results to load
time.sleep(5)

# scrape details for each search result
results = []
items = driver.find_elements(By.XPATH, '//div[@class="_2kHMtA"]')

for item in items:
    details = item.text.split('\n')
    try:
        brand = details[0].split()[0]
    except:
        brand = "-"
    try:
        name = details[0].split()[1]
    except:
        name = "-"
    try:
        color = details[0].split()[-1]
    except:
        color = "-"
    try:
        ram = details[1].split('|')[0].strip()
    except:
        ram = "-"
    try:
        rom = details[1].split('|')[1].strip()
    except:
        rom = "-"
    try:
        camera = details[2].split('|')[0].strip()
    except:
        camera = "-"
    try:
        selfie = details[2].split('|')[1].strip()
    except:
        selfie = "-"
    try:
        display = details[3].split('|')[0].strip()
    except:
        display = "-"
    try:
        battery = details[3].split('|')[1].strip()
    except:
        battery = "-"
    try:
        price = details[4]
    except:
        price = "-"
    try:
        url = item.find_element(By.XPATH, './/a').get_attribute('href')
    except:
        url = "-"
    
    results.append({
        "Brand Name": brand,
        "Smartphone name": name,
        "Colour": color,
        "RAM": ram,
        "Storage(ROM)": rom,
        "Primary Camera": camera,
        "Secondary Camera": selfie,
        "Display Size": display,
        "Battery Capacity": battery,
        "Price": price,
        "Product URL": url
    })

# save results to CSV file
with open("smartphones.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=results[0].keys())
    writer.writeheader()
    writer.writerows(results)

# quit Chrome driver
driver.quit()
```

Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# set up Chrome driver
driver = webdriver.Chrome()
driver.get("https://www.google.com/maps/")

# find search bar and search for city
search_box = driver.find_element(By.ID, "searchboxinput")
search_box.send_keys("New York City")
search_box.submit()

# wait for map to load
time.sleep(5)

# scrape latitude and longitude of city
coords = driver.current_url.split('@')[1].split(',')
lat = coords[0]
long = coords[1]

print("Latitude:", lat)
print("Longitude:", long)

# quit Chrome driver
driver.quit()
```

Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
reviews, privates from price, dorms from price, facilities and property description.

```python
import requests
from bs4 import BeautifulSoup

# URL of the page to be scraped
url = 'https://www.hostelworld.com/hostels/London'

# Send a GET request to the URL and store the response
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the hostel listings on the page
hostel_listings = soup.find_all('div', {'class': 'property-card'})

# Loop through each hostel and extract the desired data
for hostel in hostel_listings:
    # Extract hostel name
    name = hostel.find('h2', {'class': 'title'}).text.strip()
    # Extract distance from city centre
    distance_tag = hostel.find('span', {'class': 'description'})
    distance = distance_tag.text.strip() if distance_tag else ''
    # Extract overall rating
    rating_tag = hostel.find('div', {'class': 'score orange'})
    rating = rating_tag.text.strip() if rating_tag else ''
    # Extract total reviews
    reviews_tag = hostel.find('div', {'class': 'reviews'})
    reviews = reviews_tag.text.strip() if reviews_tag else ''
    # Extract price for private rooms
    private_price_tag = hostel.find('div', {'class': 'price-col privates from'})
    private_price = private_price_tag.text.strip() if private_price_tag else ''
    # Extract price for dorm rooms
    dorm_price_tag = hostel.find('div', {'class': 'price-col dorms from'})
    dorm_price = dorm_price_tag.text.strip() if dorm_price_tag else ''
    # Extract facilities
    facilities_tag = hostel.find('div', {'class': 'facilities'})
    facilities = [facility.text.strip() for facility in facilities_tag.find_all('div')] if facilities_tag else []
    # Extract property description
    description_tag = hostel.find('div', {'class': 'property-description'})
    description = description_tag.text.strip() if description_tag else ''
    
    # Print the extracted data
    print('Name:', name)
    print('Distance:', distance)
    print('Rating:', rating)
    print('Reviews:', reviews)
    print('Private from price:', private_price)
    print('Dorms from price:', dorm_price)
    print('Facilities:', facilities)
    print('Description:', description)
    print('----------------------')
```
