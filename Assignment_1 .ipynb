```python
#Write a python program to display all the header tags from wikipedia.org and make data frame.

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url)

soup = BeautifulSoup(response.content)

header_text = []

header_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

for header in header_tags:
    header_text.append(header.text)

df = pd.DataFrame({'Header Tags': header_text})

print(df)
```

2) Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
import pandas as pd 
import requests
import numpy as np
from bs4 import BeautifulSoup

url = "https://www.imdb.com/chart/top/"

page = requests.get(url)

soup = BeautifulSoup(page.content)

names = []
years = []
ratings = []


# Extraction of the movie name 
for i in soup.find_all('td',class_='titleColumn'): 
    names.append(i.text)

# Extract the movie rating
for r in soup.find_all('td', class_='ratingColumn imdbRating'):
    rating = float(r.text.strip())
    ratings.append(rating)

# Extract the movie year of release
for y in soup.find_all ('td', class_='titleColumn'):
   year = y.span.text.strip('()')
   years.append(int(year))

# Create a pandas DataFrame with the movie data
data = {'Name': names, 'Rating': ratings, 'Year': years}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# display the 50 movies by 
df.head(50)
```

Ques3) Write a python program to display IMDB’s Top rated 50 Indian movies’ data (i.e. name, rating, year of release) and make data frame.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.imdb.com/india/top-rated-indian-movies/"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

Movie_names = []
Movie_ratings = []
Movie_years = []

# Extraction of the movie name
for i in soup.find_all('td',class_='titleColumn'):
    Movie_names.append(i.text.strip())

# Extract the movie rating
for r in soup.find_all('td', class_='ratingColumn imdbRating'):
    rating = float(r.text.strip())
    Movie_ratings.append(rating)

# Extract the movie year of release
for y in soup.find_all ('td', class_='titleColumn'):
   year = y.span.text.strip('()')
   Movie_years.append(int(year))

# Create a pandas DataFrame with the movie data
data = {'Name': Movie_names, 'Rating': Movie_ratings, 'Year': Movie_years}
df = pd.DataFrame(data)

print(df)

# extracting 50 best movies 

df.head(50)
```

Ques 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) 
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://presidentofindia.nic.in/former-presidents.htm"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

Name = []

# Scraping the names of the presidents
for n in soup.find_all('div', class_='presidentListing'):
    Name.append(n.text.strip())

Term_of_office = []

# Scraping the terms of office of the presidents
for t in soup.find_all('p'):
    term_span = t.find('span', class_='terms')
    if term_span:
        duration = term_span.text.strip()
        Term_of_office.append(duration)

# Creating a DataFrame
df = pd.DataFrame({'Name': Name, 'Term of office': Term_of_office})

print(df)

```

8) Write a python program to scrape the details of most downloaded articles from AI in last 90
days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details and make data framei) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

articles = []

for article in soup.find_all('div', class_='pod-listing with-abstract'):
    title = article.find('a', class_='pod-listing-header__link').text.strip()
    authors = article.find('ul', class_='author-list').text.strip().replace('\n', ', ')
    date = article.find('span', class_='pod-listing-header__date').text.strip()
    url = article.find('a', class_='pod-listing-header__link').get('href')
    articles.append({
        'Title': title,
        'Authors': authors,
        'Published Date': date,
        'URL': url
    })

df = pd.DataFrame(articles)
print(df)
```
